# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Taking necessary columns only

```python
new_df = df[['id','name','host_id','host_name','neighbourhood_group','neighbourhood','room_type','price','minimum_nights',
             'number_of_reviews','calculated_host_listings_count','availability_365']]
new_df.head()
```

# What we learn about different host and area

```python
host_area = new_df.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
host_area.sort_values(by='calculated_host_listings_count',ascending=False).head(5)
```

# As we can see most of the listing are from Manhattan created Sonder(NYC),
# Blueground,Kara,Kazuya


# What can we learn from prediction

```python
areas_review = new_df.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
areas_review
```

```python
area = areas_review['neighbourhood_group']
review = areas_review['number_of_reviews']

fig = plt.figure(figsize=(10,5))

#Creating a bar plot between them

plt.bar(area,review,color='maroon',width=0.4)
plt.xlabel("area")
plt.ylabel("review")
plt.title("Area vs number of reviews")
plt.show()
```

```python
price_area = new_df.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area.head(5)
```

```python
area = price_area['price']
price = price_area['number_of_reviews']

fig = plt.figure(figsize=(10,6))

#creating a bar plot between them

plt.scatter(area,price)
plt.xlabel("Price")
plt.ylabel("Number Of Reviews")
plt.title("Price vs Number of Reviews")
plt.show()
```

# From the above analysis we can say that the people prefered to stay where the price is less

# Which Host are busy and why

```python
busy_host = new_df.groupby(['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
busy_host = busy_host.sort_values(by='number_of_reviews',ascending=False).head(10)
```

```python
name = busy_host['host_name']
review = busy_host['number_of_reviews']

fig = plt.figure(figsize=(10,6))

#creating a scattered plot between them
plt.bar(name,review)
plt.xlabel("Host name")
plt.ylabel("Number of review")
plt.title("Busiest Host")
plt.show()
```

# Busiest Host Are
# Dona,Ji,Maya,Carol,Danielle


# Because this host listed room as Entire home and Private room which is prefered by most number if people

# Is there any noticibel difference due to traffic in those area what could be the reason

```python
traffic_area = new_df.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_area = traffic_area.sort_values(by='minimum_nights',ascending=False)
traffic_area.head(10)
```

```python
room = traffic_area['room_type']
stayed = traffic_area['minimum_nights']

fig = plt.figure(figsize=(10,6))
#creating a bar plot betwwen them

plt.bar(room,stayed,color='yellow',width=0.4)
plt.xlabel("Room type")
plt.ylabel("Minimum Nights")
plt.title("Traffic Areas")
plt.show()
```

# From above analysis we can say that the people are very much intrested in staying in the Entire home/apt and Private room rather than Shared room
# they prefered to stay in Manhattan, Brooklyn and Queens

# **Data Preprocessing**

```python
# Checking null values
df.isnull().sum()
```

# As in our analysis the host name is not thata important so we can change them with some substitute

```python
df['name'].fillna('unknown',inplace=True)
df['host_name'].fillna('no_name',inplace=True)
```

```python
df[['host_name','name']].isnull().values.any()
```

# The column last_rewiew are having many null values so we need to drop this column

```python
df = df.drop(['last_review'],axis=1)
```

```python
df.info()
```

# Checking for duplicate values

```python
duplicate_data = df.duplicated()
print(duplicate_data.sum())
df[duplicate_data]
```

```python
df['reviews_per_month'] = df['reviews_per_month'].replace(to_replace=np.nan,value=0).astype('int64')
```

```python
df['reviews_per_month'].isnull().values.any()
```

# Now lets see the avaibility of room according to room type

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(24,12))
ax = axes.flatten()
sns.set_theme(style='white')
sns.scatterplot(data=df, x='longitude',y='latitude',hue='neighbourhood_group',ax=ax[0]);
ax[0].set_title('Location of neighbourhood groups')
sns.scatterplot(data=df[df['price']<300],x='longitude',y='latitude',hue='price',size='price',sizes=(20,60),palette='GnBu_d',ax=ax[1])
ax[1].set_title('Variation of price based on the location')
sns.scatterplot(data=df, x='longitude',y='latitude',hue='number_of_reviews',size="number_of_reviews",sizes=(20,150),palette='GnBu_d',hue_norm=(0,5),ax=ax[2])
ax[2].set_title('Variation of number of reviews based on the location')
sns.scatterplot(data=df, x='longitude',y='latitude',hue='availability_365',style='room_type',palette='GnBu_d',ax=ax[3])
ax[3].set_title('Avalability according to the Room Type')
plt.show()
```
